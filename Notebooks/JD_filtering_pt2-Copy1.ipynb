```python
import pandas as pd
import numpy as np
```

```python
new_df = pd.read_csv("Plane_Data/new_df.csv")
new_df.head(10)
```

```python
new_df["Total.Uninjured"] = new_df["Total.Uninjured"].fillna(0)
```

```python
boeing_variations = ['Boeing', 'Boeing Company', 'The Boeing Company', 'Boeing Of Canada/Dehav Div', 
                     'Boeing Company, Long Beach Div', '737', '777']

cessna_variations = ["Cessna", "Cessna Aircraft Company"]

airbus_variations = ["Airbus", "Airbus Industrie"]

piper_variations = ["Piper", "Piper Aircraft", "Piper Aircraft Corporation"]

embraer_variations = ["Embraer", "Embraer S A", "Embraer-Empresa Brasileira De", "Embraer Aircraft", "Embraer S.A."]

mcd_variations = ["Mcdonnell Douglas", "Douglas", "Mcdonnell Douglas Aircraft Co", "Mcdonnell-Douglas", 
                  "Mcdonald Douglas", "Mcdonnell Douglas Corporation"]

bombardier_variations = ["Bombardier Inc", "Bombardier", "Bombardier, Inc.", "Bombardier Learjet Corp."]

dehav_variations = ["Dehavilland", "De Havilland"]

britnor_variations = ["Britten-Norman", "Britten Norman"]

saab_variations = ["Saab-Scania Ab (Saab)", "Saab-Scania Ab", "Saab-Scania", "Saab"]

hawkbeech_variations = ["Hawker Beechcraft", "Hawker", "Hawker Beechcraft Corporation", "Hawker Beechcraft Corp.", 
                        "Beechcraft", "Hawker Siddeley", "Beech"]

gulfstream_variations = ["Gulfstream American", "Gulfstream Aerospace Lp", 
                         "Gulfstream Aerospace", "Gulfstream American Corp."]

cirrus_variations = ["Cirrus Design Corp", "Cirrus Design Corporation"]

canada_variations = ["Canadair", "Canadair Ltd"]

shortbro_variation = ["Short Bros", "Short Bros. & Harland", "Short Brothers"]
```

```python
new_df["Make"] = new_df["Make"].replace(boeing_variations, 'Boeing')
new_df["Make"] = new_df["Make"].replace(cessna_variations, 'Cessna')
```

```python
new_df["Make"].value_counts()
```

```python
new_df["Make"] = new_df["Make"].replace(airbus_variations, 'Airbus')
new_df["Make"] = new_df["Make"].replace(piper_variations, 'Piper')
new_df["Make"] = new_df["Make"].replace(embraer_variations, "Embraer")
new_df["Make"] = new_df["Make"].replace(mcd_variations, "Mcdonnell Douglas")
new_df["Make"] = new_df["Make"].replace(bombardier_variations, "Bombardier")
new_df["Make"] = new_df["Make"].replace(dehav_variations, "Dehavilland")
new_df["Make"] = new_df["Make"].replace(britnor_variations, "Britten-Norman")
new_df["Make"] = new_df["Make"].replace(saab_variations, "Saab")
new_df["Make"] = new_df["Make"].replace(hawkbeech_variations, "Hawker Beechcraft")
new_df["Make"] = new_df["Make"].replace(gulfstream_variations, "Gulfstream")
new_df["Make"] = new_df["Make"].replace(cirrus_variations, "Cirrus")
new_df["Make"] = new_df["Make"].replace(canada_variations, "Canadair")
new_df["Make"] = new_df["Make"].replace(shortbro_variation, "Short Brothers")
```

```python
val_counts = new_df["Make"].value_counts()
vals_to_keep = val_counts[val_counts > 5].index
clean_df = new_df[new_df['Make'].isin(vals_to_keep)]
```

```python
clean_df2 = clean_df["Make"].value_counts()
clean_df2
```

```python
type(clean_df2)
```

```python
# Calculate the interquartile range (IQR)
q75, q25 = np.percentile(clean_df2, [75, 25])
iqr = q75 - q25

print("Q1 (25th percentile):", q25)
print("Q3 (75th percentile):", q75)
print("Interquartile Range (IQR):", iqr)
```

```python
# Calculate the 25th percentile
threshold = clean_df["Make"].value_counts().quantile(0.25)

# Filter out rows where the "Make" value count is less than or equal to the threshold
filtered_df = clean_df.groupby("Make").filter(lambda x: len(x) > threshold)

print("Original DataFrame:")
len(clean_df)
```

```python
print("\nDataFrame after dropping values in the 25th percentile:")
len(filtered_df)
```
