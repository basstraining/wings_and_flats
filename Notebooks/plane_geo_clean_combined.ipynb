```python
import pandas as pd
```

```python
geo_df = pd.read_csv("../Data/big_plane_geo.csv")
comb_plane_df = pd.read_csv("../Data/new_comb_plane_df.csv")
```

```python
geo_df.shape
```

```python
comb_plane_df.shape
```

```python
full_df = geo_df.merge(comb_plane_df, on="Event.Id")
full_df.head(50)
```

```python
full_df.columns
```

```python
new_full_df = full_df.drop(columns=['Injury.Severity',
       'Registration.Number', 'Make:Model_x',
       'Total.Fatal.Injuries_y', 'Total.Serious.Injuries_y',
       'Total.Minor.Injuries_y', 'Total.Uninjured_y', 'extracted_value'])
```

```python
new_full_df.head()
```

```python
filepath = '../Data/comb_clean_df.csv'

# Save the new DataFrame to CSV
new_full_df.to_csv(filepath, index=False)
```

```python
new_cols = {'Total.Fatal.Injuries_x': 'Total.Fatal.Injuries', 'Total.Serious.Injuries_x': 'Total.Serious.Injuries', 'Total.Minor.Injuries_x': 'Total.Minor.Injuries', 'Total.Uninjured_x': 'Total.Uninjured', 'Make:Model_y': 'Make:Model'}
```

```python
new_full_df.rename(columns=new_cols, inplace=True)
```

```python
grouped_df = new_full_df.groupby('Make:Model').agg({
    'Severity_Score': 'mean',  # or 'sum', depending on your preference
    'Total.Fatal.Injuries': 'sum',
    'Total.Serious.Injuries': 'sum',
    'Total.Minor.Injuries': 'sum',
    'Total.Uninjured': 'sum',
    'Make:Model': 'size'
})
grouped_df.rename(columns={'Make:Model': 'Number of Events'}, inplace=True)
```
