```python
# Creating functions for uniformity among manufacturer names in seperate column

def categorize_make(make):
    if pd.isna(make):
        return "Unknown"
    elif "Boeing" in make:
        return "Boeing"
    elif "Airbus" in make:
        return "Airbus"
    elif "McdonnellDouglas" in make:
        return "Mcdonnell Douglas"
    elif "Bombardier" in make:
        return "Bombardier"
    elif "Embraer" in make:
        return "Embraer"
    elif "Saab" in make:
        return "Saab"
    elif "Fokker" in make:
        return "Fokker"
    elif "Atr" in make:
        return "Atr"
    elif "Lockheed" in make:
        return "Lockheed"
    elif "Dehavilland" in make:
        return "Dehavilland"
    elif "Short Brothers" in make:
        return "Short Brothers"  
    else:
        return "Other"
```

```python
plane_make_df['Manufacturer'] = plane_make_df['Make'].apply(categorize_make)
```

```python
manufacturer_groups = {
    'Boeing': ['Boeing'],
    'Airbus': ['Airbus'],
    'Mcdonnell Douglas': ['McdonnellDouglas'],
    'Embraer': ['Embraer'],
    'Bombardier': ['Bombardier'],
    'Lockheed': ['Lockheed'],
    'Fokker': ['Fokker'],
    'Atr': ['Atr'],
    'Short Brothers': ['Short Brothers'],
    'Dehavilland': ['Dehavilland'],
    'Unknown': ['Unknown'],
    'Saab': ['Saab'],
    }
```

```python
# Create a dictionary to store deliveries by manufacturer group
deliveries_by_group = {}

# Iterate through each plane in the total_deliveries dictionary
for plane, deliveries in total_deliveries.items():
    # Determine the manufacturer group for the current plane
    group = 'Unknown'  # Default to 'Unknown' if not found in the specified groups
    for key, values in manufacturer_groups.items():
        if any(value in plane for value in values):
            group = key
            break
    
    # If the group is not already a key in deliveries_by_group, add it with deliveries as the initial value
    if group not in deliveries_by_group:
        deliveries_by_group[group] = 0
    
    # Add the current deliveries to the total for the corresponding group
    deliveries_by_group[group] += deliveries

# Print the result
for group, total_deliveries in deliveries_by_group.items():
    print(f"{group}: {total_deliveries}")
```

```python
# Normalizing 'Severity_Score' Data

plane_make_df['Adjusted Severity_Score'] = (plane_make_df['Severity_Score'] / plane_make_df['Total Deliveries']) * 250

plane_make_df.head()
```

# Number of Incidents By Manufacturer

```python
plane_incident_data = plane_make_df.groupby('Manufacturer')[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries']].sum()


plane_melted_data = plane_incident_data.reset_index().melt(id_vars=['Manufacturer'], var_name='Injury_Type', value_name='Number_of_Injuries')

plt.figure(figsize=(14, 10))

sns.barplot(x='Manufacturer', y='Number_of_Injuries', hue='Injury_Type', data=plane_melted_data, palette='viridis')

plt.title('Number of Injuries by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Number of Injuries')

plt.xticks(rotation=45, ha='right')

plt.show()
```
